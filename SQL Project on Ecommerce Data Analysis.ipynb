### EXPLORING INSIGHTS THROUGH DATA VISUALIZATION IN THE E-COMMERCE PROJECT:

1.Number of Customers City-wise.
2.Customer Distribution by Pin Code.
3.Number of Products Category wise.
4.Stock wise Prodects.

6.Payment mode wise Total Number of Orders.

#### 1.Number of Customers City-wise. 

```python

customer_df['name'].groupby(customer_df['city']).size().plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.title('Number of Customers City-wise')
plt.show()
```

#### 2. Customer Distribution by Pin Code.

```python
customer_df['name'].groupby(customer_df['pin_code']).size().plot(kind='bar', color='purple')
plt.xlabel('PIN Code')
plt.ylabel('Number of Customers')
plt.title('Number of Customers PIN Code-wise')
plt.show()
```

#### 3.Number of Products Category wise.

```python
product_df['product_name'].groupby(product_df['category']).size().plot(kind='bar',color='yellow')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.title('Category-wise Products')
plt.show()
```

#### 4.Stock wise Prodects.

```python
plt.plot(product_df['product_name'], product_df['stock'], marker='o', linestyle='-', color='green')
plt.xlabel('Product Name')
plt.ylabel('Stock')
plt.title('Stock-wise Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 5.Order Wise Distribution.

```python
order_status_count = order_df['order_status'].value_counts()
plt.pie(order_status_count, labels=order_status_count.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightskyblue','pink'])
plt.title('Order Wise Distribution')
plt.axis('equal')
plt.show()
```

#### 6.Payment mode wise Total Number of Orders.

```python
order_df['order_id'].groupby(order_df['payment_mode']).size().plot(kind='bar',color=['red','pink','coral'])
plt.xlabel('Payment Mode')
plt.ylabel('Number of orders')
plt.title('Payment Mode-wise Total Number of Orders')
plt.show()
```

#### 7.Selling Products Based on the Quantity Sold.

```python
product_quantity_sold = product_df.groupby('product_name')['stock'].sum()
plt.pie(product_quantity_sold, labels=product_quantity_sold.index, autopct='%1.1f%%', startangle=90)
plt.title('Product Distribution Based on Quantity Sold')
plt.axis('equal') 
plt.show()
```

#### 8.Best Selling Products.

```python
best_selling_products = product_df.groupby('product_name')['stock'].sum().sort_values(ascending=False)
best_selling_products.plot(kind='barh', color='lightblue')
plt.title('Best Selling Products')
plt.ylabel('Product Name')
plt.xlabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 9.The Average, Maximum and Minimum Prices for Products.

```python
average_price=product_df['selling_price'].mean()
maximum_price=product_df['selling_price'].max()
minimum_price=product_df['selling_price'].min()
plt.bar(['Average', 'Maximum', 'Minimum'], [average_price, maximum_price, minimum_price], color=['yellow', 'green', 'red'])
plt.title('Average, Maximum, and Minimum Selling Prices for Products')
plt.ylabel('Selling Price')
plt.show()
```

#### 10.Top 3 Most Profitable Products.

```python
product_df['profit'] = product_df['selling_price'] - product_df['original_price']
top_3_profitable_products = product_df.groupby('product_name')['profit'].sum().nlargest(3)
top_3_profitable_products.plot(kind='bar', color=['lightcoral','lavender','lightgreen'])
plt.title('Top 3 Most Profitable Products')
plt.xlabel('Product Name')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

#### 11. Total Revenue Percentage for Each Category.

```python
product_df['revenue'] = product_df['stock'] * product_df['selling_price']
total_revenue_per_category = product_df.groupby('category')['revenue'].sum()
plt.pie(total_revenue_per_category, labels=total_revenue_per_category.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Revenue Percentage for Each Category')
plt.axis('equal')
plt.show()
```
